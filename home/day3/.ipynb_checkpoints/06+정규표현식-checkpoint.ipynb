```python
# 정규표현식 ( Regular Expression )

# "안수\w"
import re 
```

```python
text = """
옛날 옛적에 안수찬 형제가 살았습니다. 3형제가 있는 안수찬, 안수창, 안수빈 이렇게 3명 있었습니다.
"""

# 안수___
```

```python
# 핸드폰 번호에 대한 파싱 

# 010-2220-5736
# 010-둘둘둘영-5736
# 공일공 2220 573육       # 99.5%
```

```python
# \w ( 1 char )
# \d ( 1 decimal )
# \s ( 1 space )

# + ( 1, ..., N )
# ? ( 0, 1 )
# * ( 0, 1, .. N )

# \d{N} ( 숫자가 N개 나온다. )
# \d{N,M} ( 숫자가 N~M개 나온다 )
```

```python
# ...fly => dragonfly, butterfly
```

```python
pattern = re.compile("\d{2,3}[ -\.]?\d{3,4}[ -\.]?\d{4}")
```

```python
# \w ( 1 char )
# \d ( 1 decimal )
# \s ( 1 space )

# + ( 1, ..., N )
# ? ( 0, 1 )
# * ( 0, 1, .. N )

# \d{N} ( 숫자가 N개 나온다. )
# \d{N,M} ( 숫자가 N~M개 나온다 )
```

```python
# 주민등록번호 예제 

text = """
    안수찬 890223-1704213
    안수빈 930223-1704217
    안수창 890223-1704213      # 숫자 6자리 - 숫자 7자리 
"""
```

```python
# 이 데이터에서 주민등록번호만 찾아서, 뒷자리를 암호화(*******)
```

```python
pattern = re.compile("\d{6}-\d{7}")
# 정규표현식 GROUP
# 1. 생년월일 그룹 <birth>
# 2. 주민등록번호 뒷자리 그룹 <secret>
```

```python
pattern.findall(text)
```

```python
pattern = re.compile("(t{7})")

# 안수찬 890223-******* => 안수찬(890223-*******)
```

```python
pattern.findall(text)
```

```python
result = pattern.sub("\g<birth>-*******", text)
print(result)
```

```python
print(text)        # 안수찬(890223-*******)
```

```python
pattern = re.compile("(?P<name>\w{3}) (?P<birth>\d{6})-(?P<secret>\d{7})")
```

```python
result = pattern.sub("\g<name>(\g<birth>-*******)", text)
print(result)
```
